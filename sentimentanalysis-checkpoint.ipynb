```python
#user-defined function
def remove_pattern(text, pattern):
  result = re.findall(pattern, text)
  for word in result:
        text = re.sub(word, "", text)
  return text
```

```python
#Removing Twitter handles

combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tweet'], "@[\w]*")

#Removing URLS

combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "https://t.co/[\w]*")

#removing unicodes in ISO format. "<U+001F606> represents an emoji"
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "<[\w]*")

plus = ["+"]
for char in plus:
    combi['tidy_tweet'] = combi['tidy_tweet'].str.replace(char, ' ')

#Removing more unicode characters
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "00[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "0[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "1[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "2[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "> [\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], ">[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "&a[\w]*")

#Removing Unrelated hashtags

combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "#focusonb[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "#freemn[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "#tableshake[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "#biafr[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "#sp[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "#bbcigb[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "#oxy[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "#obi[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "#wap[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "#buhariati[\w]*")
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tidy_tweet'], "#expos[\w]*")

#Previewing the first 5 tweets
combi.head()
```

```python
#Turning all to lowercase
combi["tidy_tweet"] =combi.tidy_tweet.str.lower()

# remove special characters, numbers and punctuations
combi['tidy_tweet'] = combi['tidy_tweet'].str.replace("[^a-zA-Z#]", " ")
```

```python
#Calculating Negative, Positive, Neutral and Compound values

combi[['polarity', 'subjectivity']] = combi['tidy_tweet'].apply(lambda Text: pd.Series(TextBlob(Text).sentiment))
for index, row in combi['tidy_tweet'].iteritems():
  score = SentimentIntensityAnalyzer().polarity_scores(row)
  neg = score['neg']
  neu = score['neu']
  pos = score['pos']
  comp = score['compound']
  if neg > pos:
    combi.loc[index, 'sentiment'] = "negative"
  elif pos > neg:
    combi.loc[index, 'sentiment'] = "positive"
  else:
    combi.loc[index, 'sentiment'] = "neutral"
    combi.loc[index, 'neg'] = neg
    combi.loc[index, 'neu'] = neu
    combi.loc[index, 'pos'] = pos
    combi.loc[index, 'compound'] = comp
combi.head(10)
```

```python
#Removing Stop words

combi['tidy_tweet'] = combi['tidy_tweet'].apply(lambda x: " ".join([w for w in x.split() if len(w)>3]))

#Creating new data frames for all sentiments (positive, negative and neutral)

combi_negative = combi[combi["sentiment"]=="negative"]
combi_positive = combi[combi["sentiment"]=="positive"]
combi_neutral = combi[combi["sentiment"]=="neutral"]

def count_values_in_column(data,feature):
 total=data.loc[:,feature].value_counts(dropna=False)
 percentage=round(data.loc[:,feature].value_counts(dropna=False,
                                                   normalize=True)*100,2)
 return pd.concat([total,percentage],axis=1,keys=['Total','Percentage'])
 
#Count_values for sentiment
count_values_in_column(combi,"sentiment")
```

```python
tokenized_tweet = combi['tidy_tweet'].apply(lambda x: x.split()) #Tokenzation

#lemmatization
from nltk.stem import WordNetLemmatizer
def get_lemmatized_text(corpus):
    lemmatizer = WordNetLemmatizer()
    return [' '.join([lemmatizer.lemmatize(word) for word in test.split()]) for 
            test in corpus]

#Joining tokens back to form sentences
combi['lemma'] = get_lemmatized_text(combi['tidy_tweet'])
combi.head()
```

```python
#Hashtag Visualization
#user-defined loop function to collect hashtags

def hashtag_extract(tweets):
    hashtags = []
    for tweet in tweets:
        ht = re.findall(r"#(\w+)", tweet)
        hashtags.append(ht)
    return hashtags 

# extract hashtags from neutral tweets
ht_neutral = hashtag_extract(combi['lemma'][combi['sentiment']=="neutral"])

# extract hashtags from negative tweets
ht_negative = hashtag_extract(combi['lemma'][combi['sentiment']=="negative"])

# extract hashtags from negative tweets
ht_positive = hashtag_extract(combi['lemma'][combi['sentiment']=="positive"])
```

```python
#Word2Vec Embeddings

tokenized_tweet = combi['lemma'].apply(lambda x: x.split())
model_w2v = gensim.models.Word2Vec(
 tokenized_tweet,
 size=200, # desired no. of features/independent variables
 window=5, # context window size
 min_count=2,
 sg = 1, # 1 for skip-gram model
 hs = 0,
 negative = 10, # for negative sampling
 workers= 2, # no. of cores
 seed = 34)
model_w2v.train(tokenized_tweet, total_examples= len(combi['lemma']), 
                epochs = 20)
```

```python
def word_vector(tokens, size):
  vec = np.zeros(size).reshape((1, size))
  count = 0.
  for word in tokens:
    try:
      vec += model_w2v[word].reshape((1, size))
      count += 1.
    except KeyError: #Handling case
    
      continue
  if count != 0:
    vec /= count
    return vec
```

```python
#preparing Word2vec feature set

wordvec_arrays = np.zeros((len(tokenized_tweet), 200))
for i in range(len(tokenized_tweet)):
  wordvec_arrays[i,:] = word_vector(tokenized_tweet[i], 200)

wordvec_df = pd.DataFrame(wordvec_arrays)
wordvec_df.shape
```

Doc2Vec Embeddings

```python
#Using Gensims LabeledSentence() function to Label the tweets
def add_label(twt):
    output = []
    for i, s in zip(twt.index, twt):
      output.append(LabeledSentence(s, ["tweet_" + str(i)]))
    return output
```

```python
labeled_tweets = add_label(tokenized_tweet) # label all the tweets
```

```python
labeled_tweets[:6]
```

```python
model_d2v = gensim.models.Doc2Vec(dm=1,
                                  dm_mean=1, 
                                  size=200, 
                                  window=5,
                                  negative=7,
                                  min_count=5,
                                  workers=3,
                                  alpha=0.1,
                                  seed = 23)
model_d2v.build_vocab([i for i in tqdm(labeled_tweets)])
```

```python
model_d2v.train(labeled_tweets, total_examples= len(combi['lemma']), 
                epochs= 15)
```
